<h4>Setup</h4>

```python
git config --global user.name "Iliyana"
git config --global user.email "iliyana123@example.com"
```

<h4>Initializing repo</h4>

```python
# creates a new git repository in your folder to track all changes
git init 
```

<h4>Add files to staging area</h4>

```python
git add filename.extension      # adds specific file to staging area
git add .                       # adds ALL files in the folder to staging area
```

<h4>Commit changes</h4>

```python
git commit -m "initial commit hehehehe"          # commits the changes in the staging area with a message
```

<h4>Link to remote (github)</h4>

```python
# create a repo in your github account first, then link it here:
git remote add origin https://github.com/iliyxna/your-repo.git
```

<h4>Push to github</h4>

```python
git branch -M main                     # sets the main branch to 'main' (default is master) 
git push -u origin main                # pushes the changes to github
```

<h4>Common commands - updates</h4>

```python
git status                             # check status of files (staged, unstaged, untracked)
git log                                # view commit history
git pull                               # pulls changes from remote repository (IMPORTANT: do this before pushing changes to avoid conflicts)
git push                               # pushes changes to remote repository
```

<h4>Branching</h4> 

```python
git branch feature-branch          # create a new branch called 'feature-branch'
git checkout feature-branch        # switch to the 'feature-branch'

# make changes in this branch, then add and commit
git add .
git commit -m "added new feature"
git checkout main                 # switch back to main branch
git merge feature-branch         # merge changes from feature-branch into main

# you can delete the branch after merging IF no longer needed
git branch -d feature-branch      # delete the 'feature-branch'
```

<h4>Undo mistakes - basics</h4> 

```python
git restore filename.extension          # undo changes in working directory for specific file
git reset --hard HEAD                   # discard all changes in working directory since last commit

# these are basics for mistakes; more complex undoing may require different commands (even im not sure myself yet hehe)
```

<h4>Flow</h4> 

```python
# recap: flow
# make changes -> git add -> git commit -> git pull -> git push
git pull        # before git push to avoid conflicts
git add .
git commit -m "your message"
git push
```
